```
df = pd.read_csv("chat.csv", index_col=0)
```

```
df.head()
```

```
#change key column format to str
df["key"] = df["key"].astype(str)
```

```
text = " ".join(comment for comment in df.key)
print ("There are {} words in the combination of all comments.".format(len(text)))
```

```
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["fuck", "fucking", "shit", "flavor", "flavors"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=1600, height=400, collocations=False).generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
#Download worldcloud image from Colam
wordcloud.to_file("wc_dota.png")

from google.colab import files
files.download('wc_dota.png')
```
